```python
crossfitDF=pd.read_csv("women_open_21.1.csv")
crossfitDF
```

```python
crossfitDF.info()
```

```python
del crossfitDF["Unnamed: 0"]
```

```python
crossfitDF.rename(columns={
    '21.1': '21.1 Scores',
    '21.2': '21.2 Scores',
    '21.3': '21.3 Scores',
    '21.4': '21.4 Scores'
}, inplace=True)
```

```python
crossfitDF['first_name'] = crossfitDF['first_name'].str.title()
crossfitDF['lastname'] = crossfitDF['lastname'].str.title()
```

```python
crossfitDF.head()
```

```python
crossfitDF[['Height', 'Weight']] = crossfitDF['Height and Weight'].str.split("|", expand = True)
crossfitDF.head()
```

```python
#del crossfitDF['Height and Weight']

crossfitDF.rename(columns={
    'first_name': 'First Name',
    'lastname': 'Last Name',
    'country': 'Country',
    'continent': 'Continent',
    'affiliate': 'Affiliate',
    'points': 'Points',
    '21.1 time': '21.1 Reps',
    '21.2 time': '21.2 Reps',
    '21.3 time': '21.3 Reps',
    'weight': 'Weight Lifted'
}, inplace=True)
crossfitDF.head()
```

```python
#del crossfitDF['Height and Weight']
#crossfitDF['Age'] = crossfitDF['Age'].str.replace('Age ', '')
#crossfitDF['Weight Lifted'] = crossfitDF['Weight Lifted'].str.replace('Weight lifted: ', '')
crossfitDF['21.1 Reps'] = crossfitDF['21.1 Reps'].str.replace("reps", '')
crossfitDF['21.2 Reps'] = crossfitDF['21.2 Reps'].str.replace("reps", '')
crossfitDF['21.3 Reps'] = crossfitDF['21.3 Reps'].str.replace("reps", '')
crossfitDF.head()
```

```python
crossfitDF.info()
```

```python
crossfitDF['Age'] = crossfitDF['Age'].str.replace("Age", '')
crossfitDF['Age'] = crossfitDF['Age'].astype(int)
crossfitDF.info()
```

```python
crossfitDF = crossfitDF.dropna(how='any')
```

```python
crossfitDF.head(10)
crossfitDF.tail(10)
```

```python
#crossfitDF['21.3 Reps'] = crossfitDF['21.3 Reps'].str.replace("1 front squat", '')
# ?? 
crossfitDF2 = crossfitDF[crossfitDF['21.3 Reps'].str.contains("1 front squat")==False]
crossfitDF2
```

```python
db = mongo.CrossFitDB
CrossFitData = db.CrossFitData
```

```python
CrossFitData.insert_many(crossfitDicts)
```
